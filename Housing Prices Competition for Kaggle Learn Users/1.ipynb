```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
```

```python
data = pd.read_csv("train.csv")
```

```python
data.fillna(0,  inplace=True)
```

```python
convert = list(data.select_dtypes(['object']).columns)
```

```python
for el in convert:
    data[el] = data[el].apply(hash)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(axis=1, columns='SalePrice'), data.SalePrice, random_state=42)
```

```python
data
```

```python
from sklearn.metrics import mean_absolute_error
cat = CatBoostRegressor(silent=True)
cat.fit(X_train, y_train)
mean_absolute_error(cat.predict(X_test), y_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_validate
cat = CatBoostRegressor(silent=True, iterations=1000)

X, y = data.drop(axis=1, columns='SalePrice'), data.SalePrice
cat.fit(X, y)
cross_validate(cat, X, y, cv=10)
```

```python
test = pd.read_csv("test.csv")
test.fillna(0,  inplace=True)
convert = list(test.select_dtypes(['object']).columns)
for el in convert:
    test[el] = test[el].apply(hash)
```

```python
test
```

```python
X_test = test
X_test
```

```python
test_preds = cat.predict(X_test)
output = pd.DataFrame({'Id': X_test.Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```
